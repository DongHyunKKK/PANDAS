```python
# 년도 별로 전출전입수로 구성된 데이터프레임 생성 ==> ex) df1970, df1971, df1972, ......
num_to_seoul = []
num_to_jeju = [] 
for years in migrationDF.columns:
    print(years)
    df = pd.DataFrame(columns = admin_dis, index = admin_dis)
    for i, j in migrationDF.index:
        if i != '전출지별' and j != '전입지별':
            df.loc[i, j] = migrationDF[years][(j, i)]
        if i == j:
            df.loc[i, j] = 0
    df = df.drop('전국', axis = 1)
    df = df.T
    df = df.drop('전국', axis = 1)
    df.replace('-', 0, inplace = True)
    num_to_seoul.append(max(df.서울특별시.values))
    num_to_jeju.append(max(df.제주특별자치도.values))
    globals()['df{}'.format(years)] = df
    print(df)
    print()
```
