[문1] 아래 조건을 만족하는 Series를 생성하세요.
1) 데이터 : 1~50 범위에서 7의 배수 정수
2) 인덱스 : 'a~z' 범위에서 데이터 갯수 만큼의 알파벳
3) 타  입 : 부호가 없는 1바이트 정수

5) 속  성 : index, values, dtype, ndim, shape, size

```python
import pandas as pd
```

```python
data = range(1, 50, 7)
```

```python
sevens = pd.Series(data, index = ['a', 'b', 'c', 'd', 'e', 'f', 'g'])
sevens
```

```python
print(f'인덱스 : {sevens.index}\n값 : {sevens.values}\n데이터 타입 : {sevens.dtype}\nndim : {sevens.ndim}\nshape : {sevens.shape}\nsize : {sevens.size}')
```

2) 년도별 인구가 가장 많은 지역과 가장 적은 지역 
3) 6대 광역시에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 
4) 8도 중 7도에서 년도별 인구가 가장 많은 지역과 가장 적은 지역 

```python
file = '../DATA/stats_100701.xlsx'
```

```python
podf = pd.read_excel(file, header = 1, index_col = 0)
podf
```

```python
podf = podf.drop('계')
podf = podf.drop('수도권')
podf
```

```python
for year in [2015, 2016, 2017, 2018]:
    print(year)
    print()
    for key, value in dict(podf[year]).items():
        if value == max(podf[year].values):
            print(f'가장 많은 지역 : {key}')
        if value == min(podf[year].values):
            print(f'가장 적은 지역 : {key}')
    print()

    #print(f'가장 많은 지역 : {max(podf[year])}, 가장 적은 지역 : {min(podf[year].sort_values())}') 
```

```python
podf.idxmax()
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
file = '../DATA/stock price.xlsx'
```

```python
stockdf = pd.read_excel(file, index_col = 0)
stockdf
```

```python
stockdf.isna().sum()
```

```python
# (가)
stockdf1 = stockdf.dropna()
stockdf1[stockdf1['value'] > 10000]
stockdf1
```

```python
# (나) 이전 값으로 결측치 대체
stockdf.ffill()
```

```python
# (나) 이후 값으로 결측치 대체
stockdf.bfill()
```

```python
# (나) 평균값으로 결측치 대체

stockdf.fillna(value = {'value' : stockdf.value.mean(), 'price' : stockdf.value.mean()})
```
