## (1) 모듈 로딩

## (2) 데이터 준비

```python
dataFile = '../DATA/movies.csv'
```

## (3) 데이터 저장

```python
movieDF = read_csv(dataFile)
```

```python
movieDF
```

## (4) DataFrame의 속성들 출력

```python
for attri in movieDF.columns:
    print(attri)
movieDF.index, movieDF.values, movieDF.shape, movieDF.ndim, type(movieDF)
```

## (5) DataFrame의 컬럼별로 따로 저장하기

```python
s1 = movieDF['Rank']
s2 = movieDF['Title']
s3 = movieDF['Studio']
s4 = movieDF['Gross']
s5 = movieDF['Year']
```

## (6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, size, type 정보 출력

```python
attri_list = [s1, s2, s3, s4, s5]
attri_name = list(movieDF.columns)
# 컬럼별 values, ndim, shape, size
for a in attri_list:
    print(f'{a}의 values : {a.values}, shape : {a.shape}, ndim : {a.ndim}, size : {a.shape[0]}', sep = '\n')

# # 컬럼별 원소들 index, type
for i in range(movieDF.shape[1]):
    print(f'{attri_name[i]}')
    for j in range(movieDF.shape[0]):
        print(f'{movieDF.iloc[j][i]}의 index : {(j, i)}, type : {type(movieDF.iloc[j][i])}')
    print()
    print()
    print()
```

## (7) 컬럼별 인덱스를 변경 실습

```python
s1.index = range(1, 1 + len(s1))
s1
```

## (8) 컬럼별 데이터 즉 값 변경

```python
# s2 series를 변경
for i in range(1, len(movieDF)):
    s2[i] = '변경'
s2
```

## (9) DataFrame의 원소를 모두 출력 단, 2가지 방법 사용

```python
# (1)
for row in range(movieDF.shape[0]):
    for col in range(movieDF.shape[1]):
        print(row, col, movieDF.iloc[row, col])

print('*'*100)

for row in range(movieDF.shape[0]):
    for col in list(movieDF.columns):
        print(row, col, movieDF.loc[row, col])
```
